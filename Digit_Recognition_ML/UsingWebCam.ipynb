```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Load your trained model
model = load_model('digit_model.keras')  # Use your model file here

# Define image size (28x28 for MNIST)
IMG_SIZE = 28

# Open webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Flip the frame (optional)
    frame = cv2.flip(frame, 1)

    # Draw a rectangle as ROI (Region of Interest) for digit input
    x1, y1, x2, y2 = 200, 100, 400, 300
    cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
    roi = frame[y1:y2, x1:x2]

    # Preprocess ROI for prediction
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    thresh = cv2.threshold(blur, 100, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

    # Resize to 28x28 and reshape
    resized = cv2.resize(thresh, (IMG_SIZE, IMG_SIZE))
    normalized = resized.astype("float32") / 255.0
    reshaped = normalized.reshape(1, IMG_SIZE, IMG_SIZE, 1)

    # Predict
    prediction = model.predict(reshaped)
    digit = np.argmax(prediction)

    # Display result
    cv2.putText(frame, f'Prediction: {digit}', (x1, y1 - 10), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)

    # Show frames
    cv2.imshow("Webcam - Digit Recognition", frame)
    cv2.imshow("ROI", thresh)

    # Exit with ESC
    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
